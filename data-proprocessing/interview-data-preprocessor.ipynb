```python
import pandas as pd
import boto3
from io import StringIO
```

```python
# Initialize s3 client
s3_client = boto3.client('s3')

# Get object from s3 bucket
response = s3_client.get_object(Bucket="sagemaker-us-east-1-237863594148", 
                                Key="data-preprocessing/information-extraction.csv")

# Read csv content
csv_content = response['Body'].read().decode('utf-8')

# Convert to dataframe
df = pd.read_csv(StringIO(csv_content))

df.head(10)
```

```python
# Function to clean and split strings
def clean_and_split_string(value):
    # Handle empty cells
    if pd.isna(value) or value == 'nan' or value == '':
        return []
    
    # Handle newlines and clean the string
    value = value.replace('\n', ', ')
    
    # Split by comma and clean each item
    return [item.strip() for item in value.split(',') if item.strip() and item.lower() != 'nan']
    
df.head(10)
```

```python
# Function to process ratings
def process_rating(value):
    
    if pd.isna(value) or value == 'nan' or value == '':
        return value
    
    # Split and clean the ratings
    ratings = clean_and_split_string(str(value))
    
    if not ratings:
        return value
    
    # Convert to numbers and calculate average if multiple values exist
    try:
        numeric_ratings = [float(r) for r in ratings]
        return sum(numeric_ratings) / len(numeric_ratings) if len(numeric_ratings) > 0 else value
    except ValueError:
        return value

# Process ratings column first
if 'daen_program_rating' in df.columns:
    df['daen_program_rating'] = df['daen_program_rating'].apply(process_rating)
    
df.head(10)
```

```python
# List of columns that contain comma-separated values
columns_to_split = ['job_titles', 'job_companies', 'technologies_used', 'tools_used', 
                   'daen_program_courses', 'daen_program_tools', 
                    'daen_program_technologies', 'daen_program_valuable_skills',
                    'daen_program_potential_improvement', 'course_certification_completed']

# Function to get maximum count of items in a column
def get_max_count(series):
    return max(len(clean_and_split_string(x)) for x in series)

# Function to split a value into a list of specified length
def split_to_fixed_length(value, max_length):
    items = clean_and_split_string(value)
    # Pad with none if needed
    return items + [None] * (max_length - len(items))

# Create new dataframe for results
result_df = df.copy()

# Process each column
for col in columns_to_split:
    if col in df.columns:
        # Get maximum count for this column
        max_count = get_max_count(df[col])
        print(f"Maximum items in {col}: {max_count}")
        
        # Generate new columns for this field
        for i in range(max_count):
            col_name = f"{col}_{i+1}"
            result_df[col_name] = df[col].apply(lambda x: split_to_fixed_length(x, max_count)[i])
        
        # Drop original column
        result_df = result_df.drop(columns=[col])

print("\nTransformed shape:", result_df.shape)
with pd.option_context('display.max_columns', None, 'display.max_rows', None):
    print(result_df.head(10))
result_df.head(10)

# Verify no columns contain comma-delimited values
def check_remaining_commas(df):
    issues = []
    for column in df.columns:
        mask = df[column].astype(str).str.contains(',', na=False)
        if mask.any():
            print(f"\nProblematic values in {column}:")
            problem_rows = df[mask]
            for idx, row in problem_rows.iterrows():
                print(f"Row {idx}: {row[column]}")
            issues.append(f"Column '{column}' still contains commas in rows: {list(df[mask].index)}")
    return issues

# Check for any remaining issues
issues = check_remaining_commas(result_df)
if issues:
    print("\nWarning: Found remaining comma-delimited values:")
    for issue in issues:
        print(issue)
else:
    print("\nSuccess: No comma-delimited values remain in any column")

# Display column names to see the new structure
print("\nNew column names:")
display(result_df.columns.tolist())
```

```python
# Save expanded data
result_df.to_csv('preprocessed-information-extraction.csv', index=False)
```
